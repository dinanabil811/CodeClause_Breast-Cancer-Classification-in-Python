```python
import numpy as np
import matplotlib.pyplot as plt
```

# plot sine curve

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)

plt.plot(x,y)
plt.show
```

# plot sine and cosine curves

```python
z=np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.title("Sine and Cosine curves")
plt.legend(["sine","Cosine"]) # علشان يحددلي مين فيهم ايه بديله الاسماء اللي عيزاها بترتيب الرسم
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Subplotting sine and cosine curves

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title("subplot sine curve")

plt.subplot(2,1,2)
plt.plot(x,z)
plt.title("subplot cosine curve")
```
